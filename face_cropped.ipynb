```python
from google.colab import drive
import cv2
import os

drive.mount('/content/drive')
```

```python
input_path = "/content/drive/My Drive/dataset_for_capstone/summer"
output_path = "/content/drive/My Drive/dataset_for_capstone/summer_cropped"
os.makedirs(output_path, exist_ok=True)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml")
```

```python
# 이미지 처리
for filename in os.listdir(input_path):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_path, filename)
        image = cv2.imread(img_path)

        if image is None:
            print(f"이미지 로드 실패: {filename}")
            continue

        # 흑백으로 변환 (검출용)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # 얼굴 검출
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # 얼굴이 검출되면 크롭하여 저장 (RGB로 저장)
        for i, (x, y, w, h) in enumerate(faces):
            cropped_face = image[y:y+h, x:x+w]  # 원본 이미지에서 자르기 (RGB 유지)

            # 크롭한 얼굴 저장
            cropped_filename = f"{os.path.splitext(filename)[0]}_face_{i+1}.jpg"
            cropped_path = os.path.join(output_path, cropped_filename)
            cv2.imwrite(cropped_path, cropped_face)
            print(f"✅ 얼굴 검출 및 저장 완료: {cropped_filename}")

print("모든 이미지 처리 완료")
```
